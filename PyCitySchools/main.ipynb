```python
import pandas as pd
```

```python
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"
```

```python
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head(15)
```

```python
#calculate schools in total
total_schools =len(school_data_complete['school_name'].unique())
#Total number of students
total_students = len(school_data_complete['student_name'].unique())
#total budget
total_budget = sum(school_data_complete['budget'].unique())
# Average math score
avg_math = sum(school_data_complete['math_score']) / len(school_data_complete['math_score'])

#average reading score
avg_read = sum(school_data_complete['reading_score']) / len(school_data_complete['reading_score'])

```

```python
#percentage of students with a passing math score (70 or greater)
passing_math = [num for num in school_data_complete['math_score'] if num >= 70]
passing_math
pass_math_percent = (len(passing_math) / len(school_data_complete['math_score'])) * 100
pass_math_percent = round(pass_math_percent)

#percentage of students with a passing reading score (70 or greater)
passing_read = [num for num in school_data_complete['reading_score'] if num >= 70]
passing_read
pass_read_percent = (len(passing_read) / len(school_data_complete['reading_score'])) * 100
pass_read_percent = round(pass_read_percent)
```

```python
#overall passing rate
overall_pass_rate = round((pass_math_percent + pass_read_percent)/2,2)
```

```python
#Summary info
district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [avg_math],
    "Average Reading Score": [avg_read],
    "% Passing Math": [pass_math_percent],
    "% Passing Reading": [pass_read_percent],
    "Overall Pass Rate": [overall_pass_rate]
})
#add dollar signe and displace number by 2 digits
district_summary['Total Budget'] = district_summary['Total Budget'].map("${:,.2f}".format)

district_summary
```

```python
#School type and name
school_type = school_data.set_index(['school_name'])['type']
school_type
```

```python
#Total students
school_students_total = school_data_complete['school_name'].value_counts()
school_students_total
```

```python
#School budget
school_budget = school_data.groupby(['school_name'])['budget'].sum()
school_budget
```

```python
#Per Student Budget
per_student_budget = school_budget / school_students_total
per_student_budget
```

```python
#Average match score per school
school_math = school_data_complete.groupby(['school_name'])['math_score'].mean()
school_math = round(school_math,2)
school_math
```

```python
#Average reading score per school
school_read = school_data_complete.groupby(['school_name'])['reading_score'].mean()
school_read = round(school_read, 2)
school_read
```

```python
# % Passing Math
school_match_pass = school_data_complete[(school_data_complete['math_score'] >= 70)]
school_match_pass
percent_math_pass = school_match_pass.groupby(["school_name"])["student_name"].count() / school_students_total * 100
percent_math_pass
```

```python
#%Passing Reading
school_read_pass = school_data_complete[(school_data_complete['reading_score'] >= 70)]
school_read_pass
percent_read_pass = school_read_pass.groupby(["school_name"])["student_name"].count() / school_students_total * 100
percent_read_pass
```

```python
#Overall Passing Rate (Average of the above two)
school_overall_pass_rate = percent_math_pass + percent_read_pass / 2
school_overall_pass_rate = round(school_overall_pass_rate, 2)
school_overall_pass_rate
```

```python
#Dataframe showing school summary info
school_summary = pd.DataFrame({
    "School Type": school_type,
    "Total Students": school_students_total,
    "Total School Budget": school_budget,
    "Budget per Student": per_student_budget,
    "Average Math Score": school_math,
    "Average Reading Score": school_read,
    "% Passing Math": percent_math_pass,
    "% Passing Reading": percent_read_pass,
    "Overall Pass Rate": school_overall_pass_rate
})
#add dollar signe and tidy up data
school_summary['Total School Budget'] = school_summary['Total School Budget'].map("${:,}".format)
school_summary['Budget per Student'] = school_summary['Budget per Student'].map("${:,}".format)
school_summary['% Passing Reading'] = school_summary['% Passing Reading'].map("{:.1f}%".format)
school_summary['% Passing Math'] = school_summary['% Passing Math'].map("{:.1f}%".format)

school_summary
```

Scores by School Spending
 
Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
Average Math Score
Average Reading Score
% Passing Math
% Passing Reading
Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

school_data_complete["spending_bins"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"], spending_bins, labels = group_names)

#group by spending bins
groupby_spending = school_data_complete.groupby("spending_bins")

#average math score
math_average = groupby_spending["math_score"].mean()
#average reading score
reading_average = groupby_spending["reading_score"].mean()
#% passing math
percent_pass_math = school_data_complete[school_data_complete["math_score"] >=70].groupby("spending_bins")["Student ID"].count()/groupby_spending["Student ID"].count()
#% passing reading
percent_pass_read = school_data_complete[school_data_complete["reading_score"] >=70].groupby("spending_bins")["Student ID"].count()/groupby_spending["Student ID"].count()
#overall passing rate(average of above two)
overall_pass = (percent_pass_math + percent_pass_read)/2
                        
scores_by_school_spend = pd.DataFrame({
    "Average Math Score": math_average,
    "Average Reading Score": reading_average,
    "% Passing Math": percent_pass_math,
    "% Passing Reading": percent_pass_read,
    "% Overall Passing Rate": overall_pass
            
})
scores_by_school_spend['Average Math Score'] = scores_by_school_spend['Average Math Score'].map("{:.1f}".format)
scores_by_school_spend['Average Reading Score'] = scores_by_school_spend['Average Reading Score'].map("{:.1f}".format)
scores_by_school_spend['% Passing Math'] = scores_by_school_spend['% Passing Math'].map("{:.1f}".format)
scores_by_school_spend['% Passing Reading'] = scores_by_school_spend['% Passing Reading'].map("{:.1f}".format)
scores_by_school_spend['% Overall Passing Rate'] = scores_by_school_spend['% Overall Passing Rate'].map("{:.1f}".format)
            
scores_by_school_spend
```
